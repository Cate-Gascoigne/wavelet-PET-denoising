## Create wavelet approximation dataset

```python
data_loader = WaveletDataLoader(
    fulldose_dir='/media/uni/ST_1/LowPET/data/Train/nii_fulldose',
    lowdose_dirs=[
        '/media/uni/ST_1/LowPET/data/Train/nii_drf_4', 
        '/media/uni/ST_1/LowPET/data/Train/nii_drf_10', 
        '/media/uni/ST_1/LowPET/data/Train/nii_drf_20', 
        '/media/uni/ST_1/LowPET/data/Train/nii_drf_50', 
        '/media/uni/ST_1/LowPET/data/Train/nii_drf_100'
        ],
    h5_file='./test.h5',
    h5_file_shuffled='./test_shuffled.h5',
    compress_level=3
)
```

```python
# Create 1 (fulldose, lowdose) pair containing approximation coefficient for each DRF
approx_preview = data_loader.create_train_dataset(
    coeff_type='approx', 
    num_subject=1
)
```

```python
# Index schuffling of training data
data_loader.shuffle(name='train')
data_loader.shuffle(name='validation')

# Validate dataset size before and after shuffling
data_loader.check_size()
```

```python
data_loader.display(ds_preview=approx_preview)
```
